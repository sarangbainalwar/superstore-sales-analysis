```python
# Create order_month or order_year column
df['order_month']=pd.to_datetime(df['Order Date']).dt.month
df['order_year']=pd.to_datetime(df['Order Date']).dt.year
```

```python
df['order_month']
```

```python
df['order_year']
```

```python
#Use NumPy to create derived features (e.g., Profit Margin, flagging high discounts)
import numpy as np
df['profit_margin']=df['Profit']/df['Sales']
```

```python
df['profit_margin']
```

```python
df['net_profit'] = df['Profit'] - df['Discount'] * df['Sales']
```

```python
df['net_profit']
```

```python
df['profit_sign']=np.where(df['profit_margin']>0,'positive','negative')
```

```python
df['profit_sign']
```

```python
#flagging high discounts
df['high_discount']=np.where(df['Discount']>0.2,'yes','no')
```

```python
df['high_discount']
```

```python
df[['Segment', 'Sales', 'Discount', 'Profit', 'profit_margin', 'net_profit', 'profit_sign', 'high_discount']]\
    .where((df['high_discount'] == 'yes') & (df['profit_sign'] == 'negative'))\
    .dropna()
```

```python
# Exploratorty Data Analysis (EDA) using Pandas and NumPy
#Top-performing regions
df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
```

```python
#Top-performing categories
df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
```

```python
#Top-performing sub-categories
df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False)
```

```python
df.groupby(['Region', 'Category','Sub-Category'])['Sales'].sum().sort_values(ascending=False)
```

```python
# Which product categories give max profit but low sales?
df.groupby('Category').agg({'Sales':'sum','Profit':'sum'}).sort_values(by='Profit', ascending=False)
```

```python
# Segment-wise performance
df.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).sort_values(by='Profit',ascending=False)
```

```python
# plot for correlation between discount and profit
import matplotlib.pyplot as plt
df.plot.scatter(x='Discount',y='Profit', alpha=0.5, color='blue', edgecolor='black', figsize=(10, 6))
plt.title('Correlation between Discount and Profit')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.grid()
plt.show()
```

```python
# Time-based trend of orders
orders_by_month=df.groupby(['order_year','order_month'])['Order ID'].count().reset_index()
orders_by_month.columns=['order_year','order_month','order_count']
```

```python
# Plotting the trend of orders by month
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
sns.lineplot(
    data=orders_by_month,
    x='order_month',
    y='order_count',
    hue='order_year',
    marker='o'
)

plt.title('Monthly Order Count by Year')
plt.xlabel('Month')
plt.ylabel('Order Count')
plt.xticks(range(1, 13))
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Time-based trend of sales
sales_by_month=df.groupby(['order_year','order_month'])['Sales'].sum().reset_index()
sales_by_month.columns=['order_year','order_month','sales']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

sns.lineplot(
    data=sales_by_month,          # 👈 the dataframe
    x='order_month',              # column holding month numbers (1-12)
    y='sales',           # column with the counts / sales
    hue='order_year',             # column with the year → one colour per year
    marker='o'                    # optional markers on each point
)

plt.title('Monthly Sales by Year')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(range(1, 13))          # show 1-12 on the x-axis
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Analyze Return Rates and Their Impact on Profit by joining Orders and Returns sheets in excel file

df_orders = pd.read_excel(r'superstore_dataset.xls',sheet_name = "Orders")
df_returns = pd.read_excel(r'superstore_dataset.xls',sheet_name = "Returns")

order_return_merged = pd.merge(df_orders,df_returns, on = 'Order ID',how='left')
order_return_merged["Returned"]
```

```python
order_return_merged['Returned']=order_return_merged['Returned'].fillna('No')
```

```python
return_rate = order_return_merged.groupby('Category')['Returned'].apply(lambda x: (x=='Yes').mean()).reset_index(name='return_rate')
return_rate
```

```python
loss_by_category = order_return_merged[order_return_merged['Returned']=="Yes"].groupby('Category')['Profit'].sum().reset_index(name='loss')

loss_by_category
```

```python
sns.barplot(data=loss_by_category, x='Category', y='loss',hue='loss', palette='viridis')
plt.title('Loss by Category')
plt.xlabel('Category')
plt.ylabel('Loss')
plt.legend(title='loss', loc='upper left')
```

```python
sns.lineplot(data=return_rate,x='Category', y='return_rate')
plt.title('Return Rate by Category')
plt.xlabel('Category')
plt.ylabel('Return Rate')
plt.show()
```

```python
# Evaluate Regional Performance by Manager
df_manager = pd.read_excel(r'superstore_dataset.xls',sheet_name = "People")

df_order_manager = pd.merge(df_orders,df_manager,on="Region",how='left')
```

```python
df_order_manager
```

```python
profit_by_manager = df_order_manager.groupby(['Person','Region'])["Profit"].sum().reset_index().sort_values('Profit', ascending=False)
```

```python
profit_by_manager
```

```python
plt.figure(figsize=(12, 6))
plt.pie(profit_by_manager['Profit'],labels=profit_by_manager['Person'],autopct='%1.1f%%', startangle=140)
plt.title('Profit Distribution by Manager')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular.
plt.show()  # Show the pie chart
```

```python
plt.bar(profit_by_manager['Person'],profit_by_manager['Profit'],color='skyblue')
plt.title('Profit by Manager')
plt.xlabel('Manager')
plt.ylabel('Profit')
plt.tight_layout()  # Adjust layout to prevent label overlap
plt.show()  # Show the bar chart
```
