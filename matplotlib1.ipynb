```python
import matplotlib.pyplot as plt

months = [1, 2, 3, 4]
sales = [100, 150, 320, 120]

plt.plot(months, sales, color='blue', linestyle='-', linewidth=2, marker='o', label='2025 sales data')
plt.legend()
plt.title='Monthly Sales Data'  # Fixed typo: "monthluy" ➝ "Monthly Sales Data"
plt.xlabel('Month')              # Optional: adds clarity
plt.ylabel('Sales')              # Optional: adds clarity
plt.grid(color = 'red',linestyle=':',linewidth=1)                   # Optional: adds a grid for better readability
plt.xlim(1,4)
plt.ylim(0,200)
plt.xticks([1,2,3,4],['M1','m2','m3','m4'])
plt.show()
```

```python
plt.bar(months,sales)
```

```python
product=['a','b','c','d']
sales = [1000,2000,4023,302]

plt.barh(product,sales,color='orange',label='Sales 2025')
plt.xlabel('product')
plt.ylabel('sales')
plt.title='product sale comparison'
plt.legend()
plt.text
plt.show()
```

```python
plt.pie(sales,labels=product,autopct='%1.1f%%',colors=['gold','aqua','green','orange'])
plt.title='revenue contribuition by region'
plt.show()
```

```python
plt.hist(sales,bins = 3,color='purple',edgecolor = 'black')
plt.xlabel= 'score range'
plt.ylabel= 'no. of students'
plt.show()
```

```python
hours_studies = [1,2,3,4,5,6,7,8]
exam_scores = [50,48,85,75,74,64,35,99]

plt.scatter(hours_studies,exam_scores,color='green',marker='o',label='Student data')

plt.xlabel='hours studied'
plt.ylabel='exam score'
plt.title='relations'
plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 30, 40]

plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title("Sure")

plt.subplot(1, 2, 2)
plt.bar(x, y)
plt.title("Bar Chart")

plt.tight_layout()
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [10, 20, 30, 40]

fig,ax=plt.subplots(1,2,figsize=(10,5))
ax[0].plot(x,y,color='green')
ax[0].set_title('Line')

ax[1].bar(x,y,color='blue')
ax[1].set_title('bar')

fig.suptitle('Com')

plt.tight_layout()
plt.show()
```
